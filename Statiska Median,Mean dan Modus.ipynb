```
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import statistics as st

data5022 = pd.read_csv(r"tugas1.csv")


mean = data5022.mean()
print("Mean:", mean)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.grid(True)
plt.plot(data5022)
plt.show()
```

```
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import statistics as st

data5022 = pd.read_csv(r"tugas1.csv")


median = data5022.median()
print("Median:", median)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.grid(True)
plt.plot(data5022)
plt.show()
```
